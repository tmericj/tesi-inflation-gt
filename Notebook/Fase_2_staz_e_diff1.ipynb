## Test di Stazionarietà e Differenziazione prima

La procedura standard prevede i seguenti passaggi:
1) Si testa la **stazionarietà** delle serie nei loro livelli (in questo caso, le serie destagionalizzate).
2) Se risultano non stazionarie (come sospetto per gli indici GT, e come spesso accade per le serie economiche come il NIC), allora si applica la **differenza prima**.
3) Infine, si **ri-testa la stazionarietà** delle serie differenziate per confermare che siano diventate stazionarie (I(0)).

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss

# --- CONFIGURAZIONE ---
# Path ai file di input (output della Fase 1)
PATH_INPUT_DIR_FASE1 = "/Users/tommaso/Desktop/tesi-inflation-gt/Destagionalized_Indexes/dati_preparati_fase1" # La cartella dove hai salvato gli output della Fase 1
FILE_GT_DESTAG_IN = os.path.join(PATH_INPUT_DIR_FASE1, "indici_gt_destag_fase1.csv")
FILE_NIC_IN = os.path.join(PATH_INPUT_DIR_FASE1, "nic_fase1.csv")

# Path per i file di output di questa Fase 2
PATH_OUTPUT_DIR_FASE2 = "dati_preparati_fase2"
FILE_SERIE_STAZIONARIE_OUT = os.path.join(PATH_OUTPUT_DIR_FASE2, "indici_gt_nic_stazionari_fase2.csv")

# Colonne degli indici GT principali e del NIC da analizzare
# Assumiamo che i nomi delle colonne destagionalizzate finiscano con '_SA' se create dallo script precedente,
# o che siano i nomi originali se la destagionalizzazione non ha aggiunto un suffisso.
# Adattare se i nomi sono diversi nel tuo file 'indici_gt_destag_fase1.csv'.
COLONNE_GT_PRINCIPALI = ['indice_Inflazione_GT_PCA_SA', 'indice_Tematico_GT_SA']
COLONNA_NIC = 'NIC_destag_ISTAT'

SIGNIFICANCE_LEVEL = 0.05

# --- FUNZIONI AUSILIARIE ---
def carica_dati_da_fase1(path_file, nome_descrittivo="Dati"):
    """
    Carico i dati CSV salvati dalla Fase 1, impostando la prima colonna come indice Datetime.
    """
    print(f"--- Caricamento {nome_descrittivo} da: {path_file} ---")
    try:
        df = pd.read_csv(path_file, index_col=0) # La prima colonna è l'indice
        df.index = pd.to_datetime(df.index)
        df = df.asfreq('MS') # Assicuro la frequenza mensile
        print(f"{nome_descrittivo} caricati con successo. Shape: {df.shape}")
        print(f"Indice temporale: da {df.index.min()} a {df.index.max()}, Frequenza: {df.index.freqstr}")
        print("--- Fine Caricamento ---\n")
        return df
    except FileNotFoundError:
        print(f"ERRORE: File non trovato: {path_file}")
        return None
    except Exception as e:
        print(f"ERRORE durante il caricamento di {nome_descrittivo}: {e}")
        return None

def test_stazionarieta_completo(timeseries, series_name="Serie", significance_level=0.05, regression_kpss='c'):
    """
    Esegue i test ADF e KPSS su una serie temporale e stampa i risultati in modo strutturato.
    Restituisce True se la serie sembra stazionaria (ADF rifiuta H0 E KPSS non rifiuta H0), False altrimenti.
    """
    print(f"--- Test di Stazionarietà per: {series_name} (Regressione KPSS: '{regression_kpss}') ---")
    timeseries_clean = timeseries.dropna()

    if timeseries_clean.empty:
        print("Serie vuota dopo rimozione NaN. Impossibile testare.")
        return False # Non stazionaria se non testabile

    # Test ADF
    adf_stazionaria = False
    print("-> Test ADF:")
    adf_test = adfuller(timeseries_clean, autolag='AIC')
    adf_p_value = adf_test[1]
    print(f"   Statistica Test ADF: {adf_test[0]:.4f}, p-value: {adf_p_value:.4f}")
    if adf_p_value <= significance_level:
        print(f"   Risultato ADF: Probabilmente STAZIONARIA (p-value <= {significance_level}).")
        adf_stazionaria = True
    else:
        print(f"   Risultato ADF: Probabilmente NON STAZIONARIA (p-value > {significance_level}).")

    # Test KPSS
    kpss_stazionaria = False
    print(f"\n-> Test KPSS (regression='{regression_kpss}'):")
    # statsmodels < 0.12 usava 'legacy' per nlags, >=0.12 usa None per default (che è 'auto')
    # Per sicurezza, usiamo 'auto' se disponibile, altrimenti 'legacy'
    try:
        kpss_test = kpss(timeseries_clean, regression=regression_kpss, nlags="auto")
    except TypeError: # Per versioni più vecchie di statsmodels
        kpss_test = kpss(timeseries_clean, regression=regression_kpss, nlags="legacy")
        
    kpss_p_value = kpss_test[1]
    print(f"   Statistica Test KPSS: {kpss_test[0]:.4f}, p-value: {kpss_p_value:.4f} (#Lags: {kpss_test[2]})")
    # Per KPSS, H0 è stazionarietà. Rifiutiamo H0 se p-value è BASSO.
    # Quindi, se p-value è ALTO, NON rifiutiamo H0 => stazionaria.
    if kpss_p_value >= significance_level:
         print(f"   Risultato KPSS: NON si rifiuta H0. Probabilmente STAZIONARIA (p-value >= {significance_level}).")
         kpss_stazionaria = True
    else:
         print(f"   Risultato KPSS: Si rifiuta H0. Probabilmente NON STAZIONARIA (p-value < {significance_level}).")
    print("--- Fine Test Stazionarietà ---\n")
    
    # Considero la serie stazionaria se ADF indica stazionarietà E KPSS indica stazionarietà
    return adf_stazionaria and kpss_stazionaria


def plot_serie_e_differenziata(originale, differenziata, nome_serie_originale):
    """
    Visualizza la serie originale e la sua versione differenziata.
    """
    plt.figure(figsize=(14, 7))
    
    plt.subplot(2, 1, 1)
    plt.plot(originale.index, originale, label=f"Originale - {nome_serie_originale}", color='blue')
    plt.title(f"Serie Originale: {nome_serie_originale}")
    plt.legend()
    plt.grid(True, alpha=0.5)
    
    plt.subplot(2, 1, 2)
    plt.plot(differenziata.index, differenziata, label=f"Differenza Prima - {nome_serie_originale}", color='green')
    plt.title(f"Serie Differenziata (d=1): {nome_serie_originale}")
    plt.legend()
    plt.grid(True, alpha=0.5)
    
    plt.tight_layout()
    plt.show()

# --- ESECUZIONE SCRIPT FASE 2 ---
if __name__ == "__main__":
    print(">>> INIZIO SCRIPT FASE 2: Test Stazionarietà e Differenziazione <<<\n")

    # Creo la directory di output se non esiste
    os.makedirs(PATH_OUTPUT_DIR_FASE2, exist_ok=True)
    print(f"Directory di output per questa fase: '{PATH_OUTPUT_DIR_FASE2}'\n")

    # 1. Carico i dati dalla Fase 1
    df_gt = carica_dati_da_fase1(FILE_GT_DESTAG_IN, "Indici GT Destagionalizzati")
    df_nic = carica_dati_da_fase1(FILE_NIC_IN, "NIC Destagionalizzato ISTAT")

    if df_gt is None or df_nic is None:
        print("ERRORE CRITICO: Impossibile caricare i file di input dalla Fase 1. Script interrotto.")
        exit()

    # DataFrame per conservare le serie finali (stazionarie)
    df_final_stazionario = pd.DataFrame()

    # 2. Analisi, Differenziazione e Test per gli Indici GT
    print("--- Analisi Indici GT ---")
    for col_gt in COLONNE_GT_PRINCIPALI:
        if col_gt in df_gt.columns:
            print(f"\n*** Processo per: {col_gt} ***")
            serie_corrente_gt = df_gt[col_gt].dropna() # Lavoro su serie senza NaN iniziali/finali

            if serie_corrente_gt.empty:
                print(f"Serie '{col_gt}' vuota. Skippo.")
                continue

            # Testo la stazionarietà sui livelli (destagionalizzati)
            stazionaria_livelli = test_stazionarieta_completo(serie_corrente_gt,
                                                              series_name=f"{col_gt} (Livelli Destag)",
                                                              significance_level=SIGNIFICANCE_LEVEL,
                                                              regression_kpss='c') # 'c' per stazionarietà attorno a costante

            if not stazionaria_livelli:
                print(f"La serie '{col_gt}' (livelli destag) risulta NON STAZIONARIA. Applico differenza prima.")
                serie_differenziata_gt = serie_corrente_gt.diff().dropna()
                
                plot_serie_e_differenziata(serie_corrente_gt, serie_differenziata_gt, col_gt)

                # Testo la stazionarietà sulla serie differenziata
                stazionaria_diff = test_stazionarieta_completo(serie_differenziata_gt,
                                                               series_name=f"{col_gt} (Differenza Prima)",
                                                               significance_level=SIGNIFICANCE_LEVEL,
                                                               regression_kpss='c')
                if stazionaria_diff:
                    print(f"CONFERMATO: La serie '{col_gt}' (differenza prima) è STAZIONARIA.")
                    df_final_stazionario[f"{col_gt}_diff1"] = serie_differenziata_gt.reindex(df_gt.index) # Reindicizzo per allineamento
                else:
                    print(f"ATTENZIONE: La serie '{col_gt}' (differenza prima) è ancora NON STAZIONARIA. Considerare ulteriori trasformazioni (es. diff seconda) o revisione.")
                    df_final_stazionario[f"{col_gt}_diff1"] = serie_differenziata_gt.reindex(df_gt.index) # Salvo comunque per ispezione
            else:
                print(f"La serie '{col_gt}' (livelli destag) risulta GIÀ STAZIONARIA. Non è necessaria la differenziazione.")
                df_final_stazionario[col_gt] = serie_corrente_gt.reindex(df_gt.index)
        else:
            print(f"Attenzione: Colonna '{col_gt}' non trovata nel DataFrame degli indici GT.")
    
    # 3. Analisi, Differenziazione e Test per il NIC
    print(f"\n--- Analisi NIC ({COLONNA_NIC}) ---")
    if COLONNA_NIC in df_nic.columns:
        serie_nic = df_nic[COLONNA_NIC].dropna()

        if serie_nic.empty:
            print(f"Serie NIC '{COLONNA_NIC}' vuota. Skippo.")
        else:
            stazionaria_nic_livelli = test_stazionarieta_completo(serie_nic,
                                                                  series_name=f"{COLONNA_NIC} (Livelli Destag ISTAT)",
                                                                  significance_level=SIGNIFICANCE_LEVEL,
                                                                  regression_kpss='c') # o 'ct' se si sospetta trend forte
            if not stazionaria_nic_livelli:
                print(f"La serie '{COLONNA_NIC}' (livelli destag ISTAT) risulta NON STAZIONARIA. Applico differenza prima.")
                serie_differenziata_nic = serie_nic.diff().dropna()

                plot_serie_e_differenziata(serie_nic, serie_differenziata_nic, COLONNA_NIC)
                
                stazionaria_nic_diff = test_stazionarieta_completo(serie_differenziata_nic,
                                                                   series_name=f"{COLONNA_NIC} (Differenza Prima)",
                                                                   significance_level=SIGNIFICANCE_LEVEL,
                                                                   regression_kpss='c')
                if stazionaria_nic_diff:
                    print(f"CONFERMATO: La serie '{COLONNA_NIC}' (differenza prima) è STAZIONARIA.")
                    df_final_stazionario[f"{COLONNA_NIC}_diff1"] = serie_differenziata_nic.reindex(df_nic.index)
                else:
                    print(f"ATTENZIONE: La serie '{COLONNA_NIC}' (differenza prima) è ancora NON STAZIONARIA.")
                    df_final_stazionario[f"{COLONNA_NIC}_diff1"] = serie_differenziata_nic.reindex(df_nic.index)
            else:
                print(f"La serie '{COLONNA_NIC}' (livelli destag ISTAT) risulta GIÀ STAZIONARIA.")
                df_final_stazionario[COLONNA_NIC] = serie_nic.reindex(df_nic.index)
    else:
        print(f"Attenzione: Colonna '{COLONNA_NIC}' non trovata nel DataFrame del NIC.")

    # 4. Salvataggio del DataFrame con le serie finali (stazionarie)
    if not df_final_stazionario.empty:
        try:
            df_final_stazionario.to_csv(FILE_SERIE_STAZIONARIE_OUT)
            print(f"\nDataFrame con serie stazionarie (o trasformate) salvato in: {FILE_SERIE_STAZIONARIE_OUT}")
            print("Contenuto del DataFrame finale (prime 5 righe):")
            print(df_final_stazionario.head())
        except Exception as e_save:
            print(f"ERRORE durante il salvataggio del DataFrame finale: {e_save}")
    else:
        print("\nNessuna serie elaborata per il salvataggio finale.")

    print("\n>>> SCRIPT FASE 2 COMPLETATO <<<")
```
