## Google Trends Augmented Models (v3)

Questa è la terza versione della stima del Modello Aumentato *in-sample*, e quella definitiva. In questo script vengono aggiunte metriche come R-quadro (R²), R-quadro aggiustato (Adjusted R²), RMSE (Root Mean Squared Error) *in-sample* e MAE (Mean Absolute Error) *in-sample*.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import itertools
import warnings
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy.stats import jarque_bera
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pickle

# Ignoro i warning per un output più pulito
warnings.filterwarnings("ignore", category=sm.tools.sm_exceptions.ConvergenceWarning)
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# --- CONFIGURAZIONE GENERALE ---
PATH_INPUT_DIR_FASE2 = "/Users/tommaso/Desktop/tesi-inflation-gt/First_Difference_indexes/dati_preparati_fase2"
FILE_SERIE_STAZIONARIE_IN = os.path.join(PATH_INPUT_DIR_FASE2, "indici_gt_nic_stazionari_fase2.csv")

PATH_OUTPUT_MODELLI_AUG = "/Users/tommaso/Desktop/tesi-inflation-gt/SARIMAX_modelli/modelli_aumentati_gt_v3_insample_metrics"

COL_INFLAZIONE_STAZ = 'NIC_destag_ISTAT_diff1'
# Nomi delle colonne GT stazionarie (come da output Fase 2)
COL_GT_INFLAZIONE_STAZ = 'indice_Inflazione_GT_PCA_SA_diff1'
COL_GT_TEMATICO_STAZ = 'indice_Tematico_GT_SA_diff1'

# Dummies per outlier
DUMMY_GEN2022_DATE = '2022-01-01'
DUMMY_OTT2022_DATE = '2022-10-01'

# Ordini del modello SARIMAX base
BASE_ORDER = (1, 0, 1)
BASE_SEASONAL_ORDER = (0, 0, 0, 12)

# Parametri per la ricerca dei lag ottimali per i GT
MAX_GT_LAG_TO_TEST = 12
CRITERION_SELEZIONE_LAG = 'aic' # o 'bic'

# --- FUNZIONI AUSILIARIE (Caricamento, Creazione Dummy) ---
def carica_dati_stazionari(path_file):
    """Carico le mie serie stazionarie dal file CSV."""
    print(f"--- Caricamento Serie Stazionarie da: {path_file} ---")
    try:
        df = pd.read_csv(path_file, index_col=0)
        df.index = pd.to_datetime(df.index)
        print(f"Serie caricate. Shape: {df.shape}")
        return df
    except Exception as e:
        print(f"ERRORE caricamento: {e}")
        return None

def crea_pulse_dummy(serie_index, event_date_str, duration_months=1, name_prefix="d_pulse"):
    """Creo una pulse dummy per un evento specifico con durata personalizzabile."""
    event_ts = pd.Timestamp(event_date_str)
    date_suffix = f"{event_ts.year}_{event_ts.month:02d}"
    dummy_name = f"{name_prefix}_{date_suffix}_{duration_months}m"
    dummy = pd.Series(0, index=serie_index, name=dummy_name)
    end_event = event_ts + pd.DateOffset(months=duration_months - 1)
    event_period = serie_index[(serie_index >= event_ts) & (serie_index <= end_event)]
    if not event_period.empty:
        dummy[event_period] = 1
    # print(f"Creata dummy: {dummy_name} (attiva per {len(event_period)} periodi)") # Log ridotto
    return dummy

def calcola_metriche_insample(fitted_model, y_true_for_fit):
    """
    Calcolo R-quadro, R-quadro aggiustato, RMSE e MAE in-sample.
    y_true_for_fit deve essere la serie y usata per fittare il modello (allineata con exog).
    """
    y_pred_insample = fitted_model.fittedvalues
    
    # Rimuovo i NaN iniziali da y_true_for_fit e y_pred_insample se y_pred ha NaN all'inizio (dovuti ai lag AR/MA)
    # Questo è importante per un confronto corretto
    valid_idx = y_pred_insample.first_valid_index()
    if valid_idx is not None:
        y_true_aligned = y_true_for_fit[valid_idx:]
        y_pred_aligned = y_pred_insample[valid_idx:]
    else: # Se y_pred_insample è tutta NaN (molto improbabile per un modello fittato)
        return np.nan, np.nan, np.nan, np.nan

    if len(y_true_aligned) != len(y_pred_aligned): # Controllo di sicurezza
        print("ATTENZIONE: Lunghezze diverse tra y_true e y_pred per metriche in-sample.")
        # Provo a reindicizzare y_pred_aligned su y_true_aligned per sicurezza
        y_pred_aligned = y_pred_aligned.reindex(y_true_aligned.index)
        # Rimuovo di nuovo eventuali NaN risultanti
        common_idx = y_true_aligned.dropna().index.intersection(y_pred_aligned.dropna().index)
        y_true_aligned = y_true_aligned.loc[common_idx]
        y_pred_aligned = y_pred_aligned.loc[common_idx]


    if y_true_aligned.empty or y_pred_aligned.empty:
        return np.nan, np.nan, np.nan, np.nan


    # Per i modelli SARIMAX, l'R-quadro tradizionale può essere fuorviante.
    # Spesso non è riportato direttamente o va interpretato con cautela.
    # Infatti, fitted_model.summary() non lo include per SARIMAX.
    # Lo calcolo manualmente, consapevole che è poco usato per i modelli SARIMAX con variabili esogene.
    
    # Calcolo manuale R-quadro (come 1 - RSS/TSS)
    # RSS = somma dei quadrati dei residui
    # TSS = somma dei quadrati delle deviazioni dalla media di y_true_aligned
    rss = np.sum(fitted_model.resid[valid_idx:]**2) # Uso i residui del modello
    tss = np.sum((y_true_aligned - y_true_aligned.mean())**2)
    r_squared = 1 - (rss / tss) if tss != 0 else np.nan
    
    # R-quadro aggiustato
    n = len(y_true_aligned)
    k = len(fitted_model.params) # Numero di parametri stimati
    if n - 1 > 0 and n - k > 0 and tss != 0:
        r_squared_adj = 1 - ( (1 - r_squared) * (n - 1) / (n - k) )
    else:
        r_squared_adj = np.nan

    # RMSE e MAE sono più diretti e utili per il fit
    rmse = np.sqrt(mean_squared_error(y_true_aligned, y_pred_aligned))
    mae = mean_absolute_error(y_true_aligned, y_pred_aligned)
    
    return r_squared, r_squared_adj, rmse, mae

def stima_e_valuta_sarimax(y_series, exog_series, order, seasonal_order, model_description="Modello"):
    """Funzione robusta per stimare un SARIMAX e restituire metriche."""
    try:
        model = sm.tsa.SARIMAX(y_series, exog=exog_series, order=order, seasonal_order=seasonal_order,
                             enforce_stationarity=False, enforce_invertibility=False,
                             initialization='approximate_diffuse')
        fitted_model = None; cov_type_used = "Fallita"
        # Sequenza di tentativi per la stima con diversi tipi di covarianza
        strategies = [('HAC', {'cov_type': 'HAC', 'cov_kwds': {'kernel': 'bartlett', 'use_correction': True}}),
                      ('Robust', {'cov_type': 'robust'}), ('Default', {})]
        for name, kwargs in strategies:
            try:
                fitted_model = model.fit(disp=False, maxiter=250, **kwargs)
                cov_type_used = name; break
            except Exception: continue
        
        if fitted_model:
            aic = fitted_model.aic; bic = fitted_model.bic
            r2, r2_adj, rmse, mae = calcola_metriche_insample(fitted_model, y_series)
            print(f"  {model_description}: AIC={aic:.3f}, BIC={bic:.3f}, R2_adj={r2_adj:.3f}, RMSE={rmse:.3f}, MAE={mae:.3f}, Cov={cov_type_used}")
            return fitted_model, aic, bic, r2, r2_adj, rmse, mae, cov_type_used
    except Exception as e:
        # print(f"  ERRORE costruzione modello {model_description}: {str(e)[:70]}...") # salto il Log perché è quasi impossibile che fallisca
        pass
    return None, float('inf'), float('inf'), np.nan, np.nan, float('inf'), float('inf'), "Fallita" # se tutti i tentativi falliscono


def plot_model_diagnostics(fitted_model, model_name="SARIMAX Model", path_output_grafici=None, save_filename="diagnostics.png"):
    """Crea un pannello di diagnostiche per il modello SARIMAX stimato."""
    
    if fitted_model is None: print(f"Impossibile creare diagnostiche: modello {model_name} non valido."); return
    residuals = fitted_model.resid
    try:
        fig = fitted_model.plot_diagnostics(figsize=(15, 12))
        fig.suptitle(f'Diagnostiche Modello: {model_name}', fontsize=14, fontweight='bold', y=1.02)
        # Aggiungi linea rossa tratteggiata a y=0 solo nel primo grafico (Standardized residuals)
        fig.axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.8)    
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        if path_output_grafici and save_filename:
            full_save_path = os.path.join(path_output_grafici, save_filename)
            plt.savefig(full_save_path, dpi=300); print(f"Diagnostiche salvate in: {full_save_path}")
        plt.show()
    except Exception as e: print(f"Errore plot_diagnostics: {e}")

    print(f"\n--- Test Statistici sui Residui ({model_name}) ---")
    try:
        lb_lags = min(10, len(residuals)//5) if len(residuals)//5 > 0 else 1
        if lb_lags > 0 and len(residuals) >= lb_lags:
            lb_df = acorr_ljungbox(residuals, lags=[lb_lags], return_df=True)
            lb_pvalue = lb_df['lb_pvalue'].iloc[0] if not lb_df.empty else np.nan
            print(f"Test Ljung-Box (H0: no autocorr.): p-value ({lb_lags} lags) = {lb_pvalue:.4f}")
            if lb_pvalue < 0.05: print("  -> ATTENZIONE: Possibile autocorrelazione residua")
            else: print("  -> OK: Nessuna evidenza di autocorrelazione residua")
        else: print("Ljung-Box non eseguito (lag/dati insufficienti).")
    except Exception as e: print(f"  Errore Ljung-Box: {e}")
    try:
        if len(residuals) > 20: # Jarque-Bera per normalità
            jb_stat, jb_pvalue = jarque_bera(residuals.dropna())
            print(f"Test Jarque-Bera (H0: normalità): stat = {jb_stat:.3f}, p-value = {jb_pvalue:.4f}")
            if jb_pvalue < 0.05: print("  -> ATTENZIONE: Residui non normali")
            else: print("  -> OK: Residui normali")
        else: print("Jarque-Bera non eseguito (dati insufficienti).")
    except Exception as e: print(f"  Errore Jarque-Bera: {e}")


def confronta_modelli(risultati_modelli_list, criterio='aic'):
    """Confronta i modelli e identifica il migliore, includendo nuove metriche."""
    print(f"\n--- Confronto Modelli (Criterio di Ordinamento: {criterio.upper()}) ---")
    if not risultati_modelli_list: print("Nessun modello da confrontare."); return None
    
    # Creo DataFrame per il confronto
    # La tupla è composta da: nome modello, aic, bic, r2_adj, rmse, mae, fitted_model
    comparison_data = []
    for nome, aic, bic, r2_adj, rmse, mae, model in risultati_modelli_list:
        if model is not None:
            comparison_data.append({
                'Modello': nome,
                'AIC': aic,
                'BIC': bic,
                'R2_Adj': r2_adj,
                'RMSE_in': rmse,
                'MAE_in': mae,
                'Criterio_Ord': aic if criterio == 'aic' else bic,
                'N_Params': len(model.params) if hasattr(model, 'params') else 'N/A'
            })
    
    if not comparison_data: print("Nessun modello valido da confrontare."); return None
    
    df_comparison = pd.DataFrame(comparison_data)
    # Rimuovo la colonna 'Criterio_Ord' prima di stampare se non serve vederla
    df_comparison_sorted = df_comparison.sort_values('Criterio_Ord').drop(columns=['Criterio_Ord'])
    
    print(df_comparison_sorted.to_string(index=False, float_format='%.3f'))
    
    best_model_info = df_comparison_sorted.iloc[0]
    print(f"\nMIGLIOR MODELLO (secondo {criterio.upper()}): {best_model_info['Modello']} ({criterio.upper()}: {best_model_info[criterio.upper()]:.3f})")
    return df_comparison_sorted


# --- ESECUZIONE SCRIPT: MODELLI AUMENTATI CON GT E METRICHE IN-SAMPLE ---
if __name__ == "__main__":
    print(">>> INIZIO SCRIPT: Modelli SARIMAX Aumentati con GT (v3 - Metriche In-Sample) <<<\n")
    os.makedirs(PATH_OUTPUT_MODELLI_AUG, exist_ok=True)
    print(f"I risultati verranno salvati in: '{PATH_OUTPUT_MODELLI_AUG}'\n")

    df_serie_stazionarie = carica_dati_stazionari(FILE_SERIE_STAZIONARIE_IN)
    if df_serie_stazionarie is None: print("ERRORE CRITICO: Dati non caricati."); exit()
    if COL_INFLAZIONE_STAZ not in df_serie_stazionarie.columns:
        print(f"ERRORE: Colonna inflazione '{COL_INFLAZIONE_STAZ}' non trovata."); exit()
    y_inflazione = df_serie_stazionarie[COL_INFLAZIONE_STAZ].dropna()
    if y_inflazione.empty: print(f"ERRORE: Serie inflazione '{COL_INFLAZIONE_STAZ}' vuota."); exit()
    print(f"Serie inflazione '{y_inflazione.name}' pronta: {len(y_inflazione)} osservazioni.")

    print("\n--- Preparazione Dummies Outlier di Base ---")
    dummy_gen2022 = crea_pulse_dummy(y_inflazione.index, DUMMY_GEN2022_DATE, 1, "d_outlier")
    dummy_ott2022 = crea_pulse_dummy(y_inflazione.index, DUMMY_OTT2022_DATE, 1, "d_outlier")
    exog_base = pd.concat([dummy_gen2022.reindex(y_inflazione.index, fill_value=0),
                           dummy_ott2022.reindex(y_inflazione.index, fill_value=0)], axis=1).astype(float)
    
    risultati_tutti_modelli = [] # Lista per (nome modello, aic, bic, r2_adj, rmse, mae, fitted_model)

    # Stima del modello base (SARIMAX con solo dummy outlier) come benchmark
    print("\n--- Stima Modello Base di Riferimento (solo Dummies Outlier) ---")
    modello_base_ris, aic_base, bic_base, r2_base, r2_adj_base, rmse_base, mae_base, cov_base = stima_e_valuta_sarimax(
        y_inflazione, exog_base, BASE_ORDER, BASE_SEASONAL_ORDER, "Modello Base"
    )
    if modello_base_ris:
        risultati_tutti_modelli.append(("Modello Base", aic_base, bic_base, r2_adj_base, rmse_base, mae_base, modello_base_ris))
        # salvataggio summary e pickle
        with open(os.path.join(PATH_OUTPUT_MODELLI_AUG, "modello_base_summary.txt"), "w") as f:
             f.write(f"Modello: Modello Base\nAIC: {aic_base:.3f}, BIC: {bic_base:.3f}, R2_adj: {r2_adj_base:.3f}, RMSE: {rmse_base:.3f}, MAE: {mae_base:.3f}\nCovType: {cov_base}\n\n")
             f.write(modello_base_ris.summary().as_text())
        with open(os.path.join(PATH_OUTPUT_MODELLI_AUG, "modello_base.pkl"), "wb") as pkl_f: pickle.dump(modello_base_ris, pkl_f)
        # Diagnostiche per il modello base (in-sample fit)
        plot_model_diagnostics(modello_base_ris, model_name=f"Modello Base - Cov: {cov_base}",
                               path_output_grafici=PATH_OUTPUT_MODELLI_AUG, save_filename="diagnostica_modello_base.png")

    # Salvo il miglior lag per ogni GT per usarlo nel modello congiunto
    print("\n--- Analisi Modelli con Singolo Indicatore GT ---")
    migliori_lag_gt = {}

    for col_gt_staz in [COL_GT_INFLAZIONE_STAZ, COL_GT_TEMATICO_STAZ]:
        if col_gt_staz not in df_serie_stazionarie.columns:
            print(f"ERRORE: Colonna GT '{col_gt_staz}' non trovata."); continue
        
        print(f"\n  Valutazione per: {col_gt_staz}")
        serie_gt = df_serie_stazionarie[col_gt_staz].dropna()
        best_lag_individual = -1; best_criterion_individual = float('inf')
        temp_best_model_obj_indiv = None; temp_best_aic_indiv=float('inf'); temp_best_bic_indiv=float('inf')
        temp_best_r2_adj_indiv=np.nan; temp_best_rmse_indiv=float('inf'); temp_best_mae_indiv=float('inf')


        for k in range(1, MAX_GT_LAG_TO_TEST + 1):
            # Creo la variabile GT lagged
            gt_lagged = serie_gt.shift(k).rename(f"{col_gt_staz}_lag{k}")

            # Combino le esogene: dummy base + GT ritardato
            current_exog = pd.concat([exog_base, gt_lagged.reindex(y_inflazione.index)], axis=1).dropna()
            current_y = y_inflazione.reindex(current_exog.index)
            if len(current_y) < len(y_inflazione) * 0.8: continue # Controllo di non aver perso troppi dati

            # Stimo il modello SARIMAX con il GT lagged
            model_desc_indiv = f"Base + {gt_lagged.name}"
            fitted_model, aic, bic, r2, r2_adj, rmse, mae, _ = stima_e_valuta_sarimax(
                current_y, current_exog, BASE_ORDER, BASE_SEASONAL_ORDER, model_desc_indiv
            )
            # Se il modello è stato stimato correttamente, confronto i criteri
            if fitted_model:
                criterion_val = aic if CRITERION_SELEZIONE_LAG == 'aic' else bic
                if criterion_val < best_criterion_individual:
                    best_criterion_individual = criterion_val; best_lag_individual = k
                    temp_best_model_obj_indiv = fitted_model
                    temp_best_aic_indiv=aic; temp_best_bic_indiv=bic
                    temp_best_r2_adj_indiv=r2_adj; temp_best_rmse_indiv=rmse; temp_best_mae_indiv=mae
        
        # Dopo aver testato tutti i lag, salvo il miglior modello individuale
        if temp_best_model_obj_indiv:
            model_name_indiv = f"Base + {col_gt_staz.split('_SA_diff1')[0].split('indice_')[1]}_lag{best_lag_individual}" # Nome più pulito
            print(f"  MIGLIOR MODELLO INDIVIDUALE per {col_gt_staz}: lag={best_lag_individual}, {CRITERION_SELEZIONE_LAG.upper()}={best_criterion_individual:.3f}")
            migliori_lag_gt[col_gt_staz] = best_lag_individual # Salvo il miglior lag
            risultati_tutti_modelli.append((model_name_indiv, temp_best_aic_indiv, temp_best_bic_indiv, 
                                           temp_best_r2_adj_indiv, temp_best_rmse_indiv, temp_best_mae_indiv, 
                                           temp_best_model_obj_indiv))
            # Salvo il summary completo del miglior modello individuale
            with open(os.path.join(PATH_OUTPUT_MODELLI_AUG, f"summary_{model_name_indiv.replace(' ', '_').replace('(','').replace(')','')}.txt"), "w") as f:
                 f.write(f"Modello: {model_name_indiv}\nAIC: {temp_best_aic_indiv:.3f}, BIC: {temp_best_bic_indiv:.3f}\n")
                 f.write(f"R2_adj: {temp_best_r2_adj_indiv:.3f}, RMSE: {temp_best_rmse_indiv:.3f}, MAE: {temp_best_mae_indiv:.3f}\n")
                 f.write(f"CovType: {temp_best_model_obj_indiv.cov_type}\n\n")
                 f.write(temp_best_model_obj_indiv.summary().as_text())
            plot_model_diagnostics(temp_best_model_obj_indiv, model_name=f"{model_name_indiv} - Cov: {temp_best_model_obj_indiv.cov_type}",
                                   path_output_grafici=PATH_OUTPUT_MODELLI_AUG, save_filename=f"diagnostica_{model_name_indiv.replace(' ', '_').replace('(','').replace(')','')}.png")

    print("\n--- Stima Modello Finale Aumentato con Entrambi i GT (Migliori Lag Individuali) ---")
    exog_finale_list = [exog_base.copy()]
    desc_mod_finale = "Base"
    gt_inclusi_label = []
    
    lag_gt_i = migliori_lag_gt.get(COL_GT_INFLAZIONE_STAZ)
    if lag_gt_i:
        gt_i_lagged = df_serie_stazionarie[COL_GT_INFLAZIONE_STAZ].shift(lag_gt_i).rename(f"{COL_GT_INFLAZIONE_STAZ}_lag{lag_gt_i}")
        exog_finale_list.append(gt_i_lagged.reindex(y_inflazione.index))
        desc_mod_finale += f" + GT_Infl(lag{lag_gt_i})"
        gt_inclusi_label.append(f"GT_Infl(lag{lag_gt_i})")

    lag_gt_t = migliori_lag_gt.get(COL_GT_TEMATICO_STAZ)
    if lag_gt_t:
        gt_t_lagged = df_serie_stazionarie[COL_GT_TEMATICO_STAZ].shift(lag_gt_t).rename(f"{COL_GT_TEMATICO_STAZ}_lag{lag_gt_t}")
        exog_finale_list.append(gt_t_lagged.reindex(y_inflazione.index))
        desc_mod_finale += f" + GT_Tem(lag{lag_gt_t})"
        gt_inclusi_label.append(f"GT_Tem(lag{lag_gt_t})")

    if len(exog_finale_list) > 1: # Se almeno un GT è stato aggiunto
        exog_finale_df = pd.concat(exog_finale_list, axis=1)
        exog_finale_cleaned = exog_finale_df.dropna()
        y_finale_aligned = y_inflazione.reindex(exog_finale_cleaned.index)

        if not exog_finale_cleaned.empty and not y_finale_aligned.empty:
            mod_fin_aug_ris, aic_fin, bic_fin, r2_fin, r2_adj_fin, rmse_fin, mae_fin, cov_fin = stima_e_valuta_sarimax(
                y_finale_aligned, exog_finale_cleaned, BASE_ORDER, BASE_SEASONAL_ORDER, desc_mod_finale
            )
            if mod_fin_aug_ris:
                model_name_finale = f"Modello Aumentato ({', '.join(gt_inclusi_label)})"
                risultati_tutti_modelli.append((model_name_finale, aic_fin, bic_fin, r2_adj_fin, rmse_fin, mae_fin, mod_fin_aug_ris))
                # salvataggio summary, pickle, diagnostica e plot fit
                with open(os.path.join(PATH_OUTPUT_MODELLI_AUG, "modello_finale_augmentato_summary.txt"), "w") as f:
                    f.write(f"Modello: {model_name_finale}\nAIC: {aic_fin:.3f}, BIC: {bic_fin:.3f}, R2_adj: {r2_adj_fin:.3f}, RMSE: {rmse_fin:.3f}, MAE: {mae_fin:.3f}\nCovType: {cov_fin}\n\n")
                    f.write(mod_fin_aug_ris.summary().as_text())
                with open(os.path.join(PATH_OUTPUT_MODELLI_AUG, "modello_finale_augmentato.pkl"), "wb") as pkl_f: pickle.dump(mod_fin_aug_ris, pkl_f)
                print(f"Modello finale aumentato salvato. AIC: {aic_fin:.3f}, BIC: {bic_fin:.3f}")
                plot_model_diagnostics(mod_fin_aug_ris, model_name=f"{model_name_finale} - Cov: {cov_fin}",
                                     path_output_grafici=PATH_OUTPUT_MODELLI_AUG, save_filename="diagnostica_modello_finale_augmentato.png")
                # Plot In-sample fit
                plt.figure(figsize=(15,6)); plt.plot(y_finale_aligned.index, y_finale_aligned, label='Attuali');
                plt.plot(y_finale_aligned.index, mod_fin_aug_ris.fittedvalues, label='Fit In-Sample', color='red', linestyle='--');
                plt.title('Fit In-Sample Modello Aumentato vs Attuali'); plt.legend(); plt.grid(True, alpha=0.5);
                plt.savefig(os.path.join(PATH_OUTPUT_MODELLI_AUG, "fit_insample_modello_augmentato.png"), dpi=300); plt.show()
        else: print("Dati insufficienti per stimare il modello finale aumentato congiunto.")
    else: print("Nessun indice GT selezionato per il modello finale aumentato.")

    if risultati_tutti_modelli:
        df_confronto = confronta_modelli(risultati_tutti_modelli, CRITERION_SELEZIONE_LAG)
        if df_confronto is not None:
            confronto_file = os.path.join(PATH_OUTPUT_MODELLI_AUG, f"confronto_modelli_{CRITERION_SELEZIONE_LAG}.csv")
            df_confronto.to_csv(confronto_file, index=False)
            print(f"Confronto modelli salvato in: {confronto_file}")

    print("\n>>> SCRIPT MODELLI AUMENTATI (v3 - Metriche In-Sample) COMPLETATO <<<")
```
